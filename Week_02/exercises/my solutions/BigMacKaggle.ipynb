```python
def get_local_price(df, date, country):
    df_filtered = df.loc[(df['date'] == date) & (df['name'] == country)]
    return df_filtered['local_price'].values[0]

def calculate_price_increase(local_price_2002, local_price_2022):
    return local_price_2022 - local_price_2002

# Get local prices for Switzerland in 2002 and 2022
local_price_2002_CH = get_local_price(df, '2002-04-01', 'Switzerland')
local_price_2022_CH = get_local_price(df, '2022-07-01', 'Switzerland')

# Get local prices for the United States in 2002 and 2022
local_price_2002_USA = get_local_price(df, '2002-04-01', 'United States')
local_price_2022_USA = get_local_price(df, '2022-07-01', 'United States')

# Output local prices for Switzerland and the United States
print("Local price in 2002 (Switzerland):", local_price_2002_CH)
print("Local price in 2022 (Switzerland):", local_price_2022_CH)
print("Local price in 2002 (United States):", local_price_2002_USA)
print("Local price in 2022 (United States):", local_price_2022_USA)

# Calculate the price increase for Switzerland and the United States
price_increase_CH = calculate_price_increase(local_price_2002_CH, local_price_2022_CH)
price_increase_USA = calculate_price_increase(local_price_2002_USA, local_price_2022_USA)

# Output price increases for Switzerland and the United States
print("Price increase in Switzerland:", price_increase_CH, "CHF")
print("Price increase in United States:", price_increase_USA, "USD")


```
