```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Location': ['New York', 'Paris', 'Berlin', 'London'],
    'Age': [24, 13, 53, 33]
})
# save the DataFrame to a CSV file
df.to_csv("data/new_written_example.csv", sep=';', index=False)
```

```python
# Read TXT file

text_file = open("data/example.txt", "r")
print(text_file.read())
```

```python
# Create new and write to a new txt file
text = "Hello, this is a new text file"
with open("data/new_written_example.txt", "w") as text_file:
    text_file.write(text)
```

```python
# Create new and write to a new txt file more text

lines = ['This is a long long text', 'Wow even longer....']
with open('example_write_longtext.txt', 'w') as f:
    f.writelines(lines)
```

```python
# Read from Json
json_text = pd.read_json('data/example.json')
print(json_text)
```

```python
import pandas as pd
# Write to a new XML file
data = [
    {'name': 'Lara  Akgün', 'age': 23, 'city': 'Zürich', 'occupation': 'Software Engineer', 'hobbies': 'reading'},
    {'name': 'Pablo Akgün', 'age': 33, 'city': 'Bern', 'occupation': 'SAP specialist', 'hobbies': 'painting'}
]

df = pd.DataFrame(data)

df.to_xml('data/new_written_example.xml')
```
